This file contains visualisations for the ESLO data in the `all_selected_data` directory.

```python
# all imports
import matplotlib.pyplot as plt
```

Evolution of the distribution of interrogative structures in constructions with 'qui'.

```python
# qui in situ
x = [1972, 2014]
y = [0, 6]

x2 = [1972, 2014]
y2 = [26, 70]

for i in range(0, len(x), 2):
    plt.plot(x, y, "-b", label="Qui S in situ")
    plt.plot(x2, y2, "-b", label="Qui non-S in situ")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of qui in situ (%)')
plt.title('Qui in situ: evolution over time.')

# display the plot
plt.show()
```

Evolution of the total distribution of IM=1 (VS, est-ce que) vs IM=0 (SV).

```python
# setting of IM with quiS

x1 = [1972, 2014] # IM=1
y1 = [41, 24]

x2 = [1972, 2014] # IM=0
y2 = [59, 76]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="IM=1")
    plt.plot(x2, y2, "-r", label="IM=0")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of IM=0 and IM=1 (%)')
plt.title('Setting of IM in constructions with subject qui: evolution over time.')

# display the plot
plt.show()

# setting of IM with qui non-S

x1 = [1972, 2014] # IM=1
y1 = [46, 4]

x2 = [1972, 2014] # IM=0
y2 = [54, 96]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="IM=1")
    plt.plot(x2, y2, "-r", label="IM=0")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of IM=0 and IM=1 (%)')
plt.title('Setting of IM in constructions with non-subject qui: evolution over time.')

# display the plot
plt.show()
```

Evolution of the distribution of different interrogative strategies.

```python
# Total occurrences of different interrogative strategies

x1 = [1972, 2014] # ex situ est-ce que
y1 = [40, 19]

x2 = [1972, 2014] # ex situ VS
y2 = [1, 5]

x3 = [1972, 2014] # ex situ sv
y3 = [59, 70]

x4 = [1972, 2014] # in situ
y4 = [0, 6]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="Ex situ: est-ce que")
    plt.plot(x2, y2, "-r", label="Ex situ: VS")
    plt.plot(x3, y3, "-g", label="Ex situ: SV")
    plt.plot(x4, y4, "-y", label="In situ: SV")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences (%)')
plt.title('Interrogative strategies in constructions with subject qui: evolution over time.')

# display the plot
plt.show()

# non-subject

x1 = [1972, 2014] # ex situ est-ce que
y1 = [26, 0]

x2 = [1972, 2014] # ex situ VS
y2 = [28, 4]

x3 = [1972, 2014] # ex situ sv
y3 = [20, 26]

x4 = [1972, 2014] # in situ
y4 = [26, 70]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="Ex situ: est-ce que")
    plt.plot(x2, y2, "-r", label="Ex situ: VS")
    plt.plot(x3, y3, "-g", label="Ex situ: SV")
    plt.plot(x4, y4, "-y", label="In situ: SV")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences (%)')
plt.title('Interrogative strategies in constructions with non-subject qui: evolution over time.')

# display the plot
plt.show()
```

Evolution in the realisation of clefts.

```python
# evolution of clefts with quiS

x1 = [1972, 2014] # Cop>Wh (regular)
y1 = [26, 82]

x2 = [1972, 2014] # Wh>Cop (reversed)
y2 = [74, 18]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="Cop>Wh (regular)")
    plt.plot(x2, y2, "-r", label="Wh>Cop (reversed)")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of clefts (%)')
plt.title('Clefting in constructions with subject qui: evolution over time.')

# display the plot
plt.show()

# evolution of clefts with qui non-S

x1 = [1972, 2014] # Cop>Wh (regular)
y1 = [100, 100]

x2 = [1972, 2014] # Wh>Cop (reversed)
y2 = [0, 0]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="Cop>Wh (regular)")
    plt.plot(x2, y2, "-r", label="Wh>Cop (reversed)")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of clefts (%)')
plt.title('Clefting in constructions with non-subject qui: evolution over time.')

# display the plot
plt.show()
```
