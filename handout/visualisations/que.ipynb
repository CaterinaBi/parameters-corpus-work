```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# all lines

# create data
x = ['late 1800', 'early 1900']

# y axes
queVS = [43, 19]
quVS = [26, 13]
quest  = [31, 68]
  
# plot lines
plt.plot(x, queVS, label = "que VS", linestyle="-")
plt.plot(x, quVS, label = "qu' VS", linestyle="-")
plt.plot(x, quest, label = "qu'est-ce que", linestyle="-")
plt.legend()
plt.show()
```

```python
# que

labels = ['que VS', 'qu\' VS', 'qu\'est-ce que']
late1800 = [43, 26, 31]
early_1900 = [19, 13, 68]

x = np.arange(len(labels))  # the label locations
width = 0.20  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, late1800, width, label='late 1800')
rects2 = ax.bar(x + width/2, early_1900, width, label='early 1900')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of occurrences')
ax.set_title('Evolution of preferences for interrogatives types in the presence of que')
ax.set_xticks(x, labels)
ax.legend()

# ax.bar_label(rects1, padding=1)
# ax.bar_label(rects2, padding=1)
ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)

fig.tight_layout()

plt.show()
```
