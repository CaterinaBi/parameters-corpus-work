```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# VS

labels = ['comment', 'o√π', 'quand', 'quiO', 'quoi0']
late1800 = [97, 95, 79, 93, 83]
early_1900 = [87, 92, 75, 89, 100]
eslo1 = [25, 28, 53, 38, 22]
eslo2 = [6, 4, 8, 4, 0]

x = np.arange(len(labels))  # the label locations
width = 0.20  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*1.5, late1800, width, label='late 1800')
rects2 = ax.bar(x - width/2, early_1900, width, label='early 1900')
rects3 = ax.bar(x + width/2, eslo1, width, label='1970s')
rects4 = ax.bar(x + width*1.5, eslo2, width, label='2010s')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of occurrences')
ax.set_title('Distribution of VS constructions')
ax.set_xticks(x, labels)
ax.legend()

# ax.bar_label(rects1, padding=1)
# ax.bar_label(rects2, padding=1)
ax.bar_label(rects3, padding=1)
ax.bar_label(rects4, padding=1)

fig.tight_layout()

plt.show()
```

```python
# SV

labels = ['comment', 'o√π', 'quand', 'quiO', 'quoi0']
late1800 = [3, 5, 21, 7, 17]
early_1900 = [13, 8, 25, 11, 0]
eslo1 = [75, 72, 47, 62, 78]
eslo2 = [94, 96, 92, 96, 100]

x = np.arange(len(labels))  # the label locations
width = 0.20  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width*1.5, late1800, width, label='late 1800')
rects2 = ax.bar(x - width/2, early_1900, width, label='early 1900')
rects3 = ax.bar(x + width/2, eslo1, width, label='1970s')
rects4 = ax.bar(x + width*1.5, eslo2, width, label='2010s')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of occurrences')
ax.set_title('Distribution of SV constructions (ex situ and in situ)')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)
ax.bar_label(rects3, padding=1)
ax.bar_label(rects4, padding=1)

fig.tight_layout()

plt.show()
```
