This file contains visualisations for the ESLO data in the `all_selected_data` directory.

```python
# all imports
import matplotlib.pyplot as plt
```

Evolution of the distribution of interrogative structures in constructions with 'comment'.

```python
# comment in situ
x = [1972, 2014]
y = [4, 32]

for i in range(0, len(x), 2):
    plt.plot(x, y, "-b", label="Wh-in situ")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of wh-in situ (%)')
plt.title('Comment in situ: evolution over time.')

# display the plot
plt.show()
```

Evolution of the total distribution of IM=1 (VS, est-ce que) vs IM=0 (SV).

```python
# setting of IM with comment

x1 = [1972, 2014] # IM=1
y1 = [51, 16]

x2 = [1972, 2014] # IM=0
y2 = [49, 84]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="IM=1")
    plt.plot(x2, y2, "-r", label="IM=0")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of IM=0 and IM=1 (%)')
plt.title('Setting of IM in constructions with comment: evolution over time.')

# display the plot
plt.show()
```

Evolution of the distribution of different interrogative strategies.

```python
# Total occurrences of different interrogative strategies

x1 = [1972, 2014] # ex situ est-ce que
y1 = [52, 10]

x2 = [1972, 2014] # ex situ VS
y2 = [13, 4]

x3 = [1972, 2014] # ex situ sv
y3 = [11, 10]

x4 = [1972, 2014] # in situ
y4 = [24, 76]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="Ex situ: est-ce que")
    plt.plot(x2, y2, "-r", label="Ex situ: VS")
    plt.plot(x3, y3, "-g", label="Ex situ: SV")
    plt.plot(x4, y4, "-y", label="In situ: SV")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences (%)')
plt.title('Interrogative strategies in constructions with comment: evolution over time.')

# display the plot
plt.show()
```

Evolution in the realisation of clefts.

```python
# evolution of clefts with comment

x1 = [1972, 2014] # Cop>Wh (regular)
y1 = [66, 86]

x2 = [1972, 2014] # Wh>Cop (reversed)
y2 = [34, 14]

for i in range(0, len(x), 2):
    plt.plot(x1, y1, "-b", label="Cop>Wh (regular)")
    plt.plot(x2, y2, "-r", label="Wh>Cop (reversed)")

# define legend
plt.legend(loc="upper left")

# define labels
plt.xlabel('ESLO 1 (1972) to ESLO 2 (2014)')
plt.ylabel('Total occurrences of clefts (%)')
plt.title('Clefting in constructions with comment: evolution over time.')

# display the plot
plt.show()
```
